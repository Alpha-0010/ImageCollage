```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the images
img1=plt.imread('./img1.jpg')
img2=plt.imread('./img2.jpg')
img3=plt.imread('./img3.jpg')
img4=plt.imread('./img4.jpg')
img5=plt.imread('./img5.jpeg')
```

```python
# Resize the 4 images to fit in the frame .i.e. make them 200*200
img1=cv2.resize(img1,(200,200))
img2=cv2.resize(img2,(200,200))
img3=cv2.resize(img3,(200,200))
img4=cv2.resize(img4,(200,200))
img5=cv2.resize(img5,(100,100)) #Resize it into 100*100
```

```python
# Create a black frame to fit all the images
finalImg=np.zeros((430,430,3),dtype='uint8')
```

```python
# Now add all the pictures into the black frame.
finalImg[10:210,10:210,:]=img1
finalImg[10:210,220:420,:]=img2
finalImg[220:420,10:210,:]=img3
finalImg[220:420,220:420,:]=img4
```

```python
# Now add the fifth Image in the center with borders of length 10px.
centreImg=np.zeros((120,120,3),dtype='uint8')
```

```python
centreImg[10:110,10:110,:]=img5
```

```python
finalImg[155:275,155:275,:]=centreImg
```

```python
plt.axis('off')
plt.imshow(finalImg)
plt.show()
```

```python
plt.imsave("res.jpg",finalImg)
```
